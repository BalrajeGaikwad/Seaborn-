# Violin Plot

**A violin plot plays a similar role as a box and whisker plot . It shows the distribution of quantitative data across several levels of one (or more) categorical variables such that those distributions can be compared**

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
var=sns.load_dataset("tips")
var
```

```python
sns.violinplot(x="day", y="total_bill", data=var,hue="time")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Create a dataset: Salary distribution for different job roles
np.random.seed(42)
data = {
    "Job Role": np.random.choice(["Software Engineer", "Data Scientist", "Product Manager", "Designer"], 200),
    "Salary ($K)": np.concatenate([
        np.random.normal(110, 15, 50),  # Software Engineer salaries
        np.random.normal(120, 20, 50),  # Data Scientist salaries
        np.random.normal(130, 18, 50),  # Product Manager salaries
        np.random.normal(90, 10, 50)   # Designer salaries
    ])
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Create a Violin Plot
plt.figure(figsize=(10, 6))
sns.violinplot(x="Job Role", y="Salary ($K)", data=df, palette="coolwarm", inner="quartile")

# Title & Labels
plt.title("Salary Distribution by Job Role", fontsize=14)
plt.xlabel("Job Role")
plt.ylabel("Salary ($K)")
plt.grid(True)
plt.show()
```
